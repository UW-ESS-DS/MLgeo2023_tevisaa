```python
# Import packages here
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Generate two lists of five random integers ranging in value from 1 to 100
randomX = [random.randint(1, 100) for val in range(5)]
randomY = [random.randint(1, 100) for val in range(5)]

# Convert the list into a numpy array
npArrayX = np.array(randomX)
npArrayY = np.array(randomY)

# Divide the x values by the y values
dividedValues = npArrayX/npArrayY

# Find the norm
norm = np.linalg.norm(dividedValues)

# Normalize the data
normalizedValues = dividedValues/norm

# Create a plot
fig, ax = plt.subplots(figsize=(10, 5))

# Define plot variables
ax.plot(normalizedValues, '-')
ax.set_xlabel('X value')
ax.set_ylabel('Y value')
ax.set_title('Test Plot')
ax.grid(True)

# Save the figure
plt.savefig('testPlot.png')

# Show the plot
plt.show()
```
